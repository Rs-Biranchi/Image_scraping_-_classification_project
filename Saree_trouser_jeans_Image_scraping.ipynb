```python
from selenium import webdriver
import requests
driver=webdriver.Chrome(executable_path='chromedriver.exe')
import os
```

```python
saree_link="https://www.amazon.in/s?k=saree&ref=nb_sb_noss_2"
saree_page=driver.get(saree_link)
saree_image1=driver.find_elements_by_xpath("//img[@class='s-image']")
print("The number of Saree in page1 is ",len(saree_image1))
```

# Creating directory to Save the images

```python
def create_directory(dirname):
    current_path=os.getcwd()
    path=os.path.join(current_path,dirname)
    if not os.path.exists(dirname):
        os.makedirs(path)
        
create_directory("Saree_images")
```

```python
for index,link in enumerate(saree_urls):
    print("Downloading {0} of {1} from saree_urls".format(index+1,len(saree_urls)))
    response=requests.get(link)
    with open("Saree_images/saree{}_{}.jpeg".format(index,1),'wb') as file:
        file.write(response.content)
```

```python
create_directory("Trouser_images")

for index,link in enumerate(trouser_urls):
    print("Downloading {0} of {1} from trouser_urls".format(index+1,len(trouser_urls)))
    response=requests.get(link)
    with open("Trouser_images/trouser{}_{}.jpeg".format(index,1),'wb') as file:
        file.write(response.content)
```

```python
create_directory("Jeans_images")

for index,link in enumerate(jeans_urls):
    print("Downloading {0} of {1} from jeans_urls".format(index+1,len(jeans_urls)))
    response=requests.get(link)
    with open("Jeans_images/jeans{}_{}.jpeg".format(index,1),'wb') as file:
        file.write(response.content)
```
